## Data Analysis

```python
counts = [-0.140673, -0.246565, -0.119645, -0.250904, 0.114527, -0.344461]
thetas = [np.pi, np.pi-np.pi/8, np.pi/2-np.pi/8, np.pi/2, np.pi/2-np.pi/4, np.pi/2+np.pi/4]
thetas_fit = np.linspace(0, np.pi)
counts_fit = 0.33*np.cos(thetas_fit+np.pi/2-np.pi/4+0.2)

plt.scatter(thetas, counts)
plt.plot(thetas_fit, counts_fit)
plt.show()
```
